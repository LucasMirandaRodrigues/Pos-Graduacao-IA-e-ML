## 4 Tratamento dos Dados

#### 4.1 Dados Discrepantes

Substitui os valores discrepantes pelo valor médio.

```python
dados_tratados=dados.copy()
dados_tratados=dados_tratados.drop(['Frequência'],axis=1)

# Amplitude Pista Interna
dados_tratados.loc[dados_tratados['Amplitude Pista Interna']>LS_pi,'Amplitude Pista Interna']=analise.loc['mean','Amplitude Pista Interna']

# Amplitude Pista Externa
dados_tratados.loc[dados_tratados['Amplitude Pista Externa']>LS_pe,'Amplitude Pista Externa']=analise.loc['mean','Amplitude Pista Externa']

# Amplitude Elemento Rolante
dados_tratados.loc[dados_tratados['Amplitude Elemento Rolante']>LS_er,'Amplitude Elemento Rolante']=analise.loc['mean','Amplitude Elemento Rolante']

# X RMS
dados_tratados.loc[dados_tratados['X RMS']>LS_x,'X RMS']=analise.loc['mean','X RMS']

# Fator de Crista
dados_tratados.loc[dados_tratados['Fator de Crista']>LS_fc,'Fator de Crista']=analise.loc['mean','Fator de Crista']

# Fator K
dados_tratados.loc[dados_tratados['Fator K']>LS_fk,'Fator K']=analise.loc['mean','Fator K']

dados_tratados.head(10)
```

#### 4.3 Normalização

Normalizar os números na escala de 0 a 1.

```python
# Amplitude Pista Interna
dados_tratados['Amplitude Pista Interna']=((dados_tratados['Amplitude Pista Interna'] - dados_tratados['Amplitude Pista Interna'].min())
                                            /(dados_tratados['Amplitude Pista Interna'].max() - dados_tratados['Amplitude Pista Interna'].min()))

# Amplitude Pista Externa
dados_tratados['Amplitude Pista Externa']=((dados_tratados['Amplitude Pista Externa'] - dados_tratados['Amplitude Pista Externa'].min())
                                            /(dados_tratados['Amplitude Pista Externa'].max() -dados_tratados['Amplitude Pista Externa'].min()))

# Amplitude Elemento Rolante
dados_tratados['Amplitude Elemento Rolante']=((dados_tratados['Amplitude Elemento Rolante'] - dados_tratados['Amplitude Elemento Rolante'].min())
                                            /(dados_tratados['Amplitude Elemento Rolante'].max() - dados_tratados['Amplitude Elemento Rolante'].min()))

# X RMS
dados_tratados['X RMS']=((dados_tratados['X RMS'] - dados_tratados['X RMS'].min())
                                            /(dados_tratados['X RMS'].max() - dados_tratados['X RMS'].min()))

# Fator de Crista
dados_tratados['Fator de Crista']=((dados_tratados['Fator de Crista'] - dados_tratados['Fator de Crista'].min())
                                            /(dados_tratados['Fator de Crista'].max() - dados_tratados['Fator de Crista'].min()))

# Fator K 
dados_tratados['Fator K']=((dados_tratados['Fator K'] - dados_tratados['Fator K'].min())
                                            /(dados_tratados['Fator K'].max() - dados_tratados['Fator K'].min()))

# Carga (HP)
dados_tratados['Carga (HP)']=((dados_tratados['Carga (HP)'] - dados_tratados['Carga (HP)'].min())
                                            /(dados_tratados['Carga (HP)'].max() - dados_tratados['Carga (HP)'].min()))

# Tamanho Defeito (mm)
dados_tratados['Tamanho Defeito (mm)']=((dados_tratados['Tamanho Defeito (mm)'] - dados_tratados['Tamanho Defeito (mm)'].min())
                                            /(dados_tratados['Tamanho Defeito (mm)'].max() - dados_tratados['Tamanho Defeito (mm)'].min()))
dados_tratados.head(10)
```
